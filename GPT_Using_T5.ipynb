```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-large")
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-large")
```

**Creating a chat function inorder to receive texts from user and provide a response**

```python
def chat(text):
    input_ids = tokenizer.encode(text + '</s>', return_tensors='pt')
    output = model.generate(input_ids, max_length=1000, do_sample=True)
    response = tokenizer.decode(output[0], skip_special_tokens=True)
    return response
```

**Gradio Ui for the user**

```python
chat_input = gr.inputs.Textbox(label="Enter Text")
chat_output = gr.outputs.Textbox(label="Response")

gr.Interface(fn=chat, inputs=chat_input, outputs=chat_output).launch()
```

```python
def generate_summary(input_text):
    input_ids = tokenizer.encode(input_text, return_tensors="pt")
    summary_ids = model.generate(input_ids)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary
```

```python
input_text = gr.inputs.Textbox(lines=10, label="Input Text")
output_text = gr.outputs.Textbox(label="Output Summary")
title = "T5 large Summarization"
description = "Enter text and generate a summary using the T5 large model."
```

```python
gr.Interface(fn=generate_summary, inputs=input_text, outputs=output_text, title=title, description=description).launch()
```
